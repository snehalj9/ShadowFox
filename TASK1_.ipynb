##  Part 1 — Matplotlib
 Fine-grained control, publication-quality figures

### 1. Line Chart
Best for: trends over time, continuous data, mathematical functions

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [100, 150, 120, 200, 180, 220]

plt.figure(figsize=(8, 5))
plt.plot(months, sales, marker='o', linewidth=2, markersize=8)
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True, alpha=0.3)
plt.show()
```

### 2. Bar Chart
Best for: comparing categorical values, rankings, survey results

```python
import matplotlib.pyplot as plt

categories = ['Product A', 'Product B', 'Product C', 'Product D']
values = [50, 70, 60, 80]

plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='steelblue')
plt.title('Product Sales Comparison')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

### 3. Scatter Plot
Best for: correlation between two continuous variables, clustering, outliers

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50) * 100
y = np.random.rand(50) * 100

plt.figure(figsize=(8, 5))
plt.scatter(x, y, alpha=0.6, s=100, color='coral')
plt.title('Scatter Plot Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

### 4. Histogram
Best for: distribution of a single variable, frequency analysis

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Data')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

### 5. Pie Chart
Best for: part-to-whole proportions (use sparingly — bar charts are often clearer)

```python
import matplotlib.pyplot as plt

labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [30, 25, 20, 25]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart Example')
plt.show()
```

### 7. Heatmap
Best for: correlation matrices, confusion matrices, grid-based data

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8],
    'C': [9, 10, 11, 12]
})

plt.figure(figsize=(8, 5))
sns.heatmap(data, annot=True, cmap='YlOrRd', fmt='d')
plt.title('Heatmap Example')
plt.show()
```

### 8. Box Plot
Best for: comparing distributions across groups, spotting outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

plt.figure(figsize=(8, 5))
sns.boxplot(x='day', y='total_bill', data=tips, palette='Set2')
plt.title('Box Plot: Bill Amount by Day')
plt.show()
```

### 9. Violin Plot
Best for: showing full distribution shape across categories (richer than box plots)

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

plt.figure(figsize=(8, 5))
sns.violinplot(x='species', y='sepal_length', data=iris, palette='muted')
plt.title('Violin Plot: Sepal Length by Species')
plt.show()
```

### 10. Pair Plot
Best for: EDA — visualising all pairwise feature relationships at once

```python
import seaborn as sns

iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species', diag_kind='hist')
plt.show()
```

### 11. KDE Plot (Kernel Density Estimate)
Best for: smooth continuous distribution comparison across groups

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

plt.figure(figsize=(8, 5))
sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True)
plt.title('KDE Plot: Distribution of Bill Amount')
plt.show()
```

### 12. Regression Plot
Best for: visualising linear relationships with confidence intervals

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

plt.figure(figsize=(8, 5))
sns.regplot(x='total_bill', y='tip', data=tips, scatter_kws={'alpha':0.6})
plt.title('Regression Plot: Tip vs Bill Amount')
plt.show()
```
