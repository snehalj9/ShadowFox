**Matplotlib**

Matplotlib is a basic and widely used Python library for data visualization. It allows users to create simple graphs like line plots, bar charts, and histograms. It is highly customizable and is mainly used for creating static and detailed visualizations. It is suitable for beginners who want to understand the fundamentals of plotting.

**Seaborn**

Seaborn is a Python library built on top of Matplotlib that provides more attractive and advanced visualizations. It is easier to use and comes with built-in themes and color palettes. It is mainly used for statistical data visualization and works well with datasets. It helps in creating visually appealing graphs with less code.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```



**Line Plot :**
This graph is used to show trends over time or continuous data. It helps in understanding how values change.

Use Case:
Tracking sales, temperature changes, etc.

```python
#lineplot
import matplotlib.pyplot as plt

x = [1,2,6,9]
y = [10,17,25,35]

plt.plot(x,y)
plt.title("Line Plot")
plt.show()
```

**Bar Chart:**
This graph is used to compare different categories. Each bar represents a value.

Use Case:
Comparing marks, product sales, etc.


```python
x = ["A","B","C"]
y = [5,7,3]

plt.bar(x, y, color='pink')
plt.title("Category Comparison")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

**Scatter Plot:**
This graph shows the relationship between two variables. It helps to identify patterns or correlations.

Use Case:
Finding correlation between data

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex")
plt.title("Tips vs Total Bill")
plt.show()
```

**Histogram (Matplotlib)**

A histogram is used to represent the distribution of numerical data. It divides the data into intervals (called bins) and shows how many values fall into each interval. This helps in understanding the spread and frequency of the data.

Use Case:

Analyzing marks distribution of students ,Understanding age distribution in a group

```python
import matplotlib.pyplot as plt

data = [10,12,13,15,15,16,18,20,22,25]

plt.hist(data, bins=5, color='blue', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```

**Pie Chart (Matplotlib)**

A pie chart is used to represent data as parts of a whole. Each slice shows the proportion or percentage of a category compared to the total. It is useful for showing relative contributions.

Use Case:
Market share of companies ,Budget distribution

```python
import matplotlib.pyplot as plt

labels = ['A','B','C']
sizes = [30,40,30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart Example")
plt.show()
```
